```python
# %load import\ modules.py
#!/usr/bin/python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import *
from sklearn.cross_validation import *
```

```python
# data is heart disease data from UCI Machine Learning repositoy 
# "https://archive.ics.uci.edu/ml/datasets/Heart+Disease"
# data has been preprocessed and rows with missing values have been dropped

my_data = pd.read_csv("../data/clevelandna.data")
my_data = my_data.iloc[:,1:]
my_data.head()
```

```python
X = my_data.iloc[:,:13] # variable 13 is an output variable
y = my_data.iloc[:,14]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42)
print("training set size: " + str(len(X_train)))
print("test set size: " + str(len(X_test)))
```

```python
std_scale = preprocessing.StandardScaler().fit(X_train)
# WHAT DOES THIS DO EXACTLY??
X_train_std = std_scale.transform(X_train)
X_test_std = std_scale.transform(X_test)
```

```python
# function for retrieving roc parameters

def roc_params(estimator, train_X, test_X):
    model = estimator
    model.fit(train_X, y_train)
    test_pred = model.predict_proba(test_X)
    
    #return roc parameters
    fpr, tpr, _ = roc_curve(y_test, test_pred[:,1])
    auc = roc_auc_score(y_test, test_pred[:,1])
    roc_params = [fpr, tpr, auc]
    return roc_params
```

```python
estimators = [LogisticRegression(), SVC(probability=True), GaussianNB(), DecisionTreeClassifier(max_depth = 4), RandomForestClassifier(), KNeighborsClassifier(n_neighbors=9)]
names = ["Logistic Regression", "Support Vector Machine", "Gaussian Naive Bayes", "Decision Tree Classifier","Random Forest Classifier", "K Neighbors (9)"]
colors = ["b", "k", "g", "r", "c", "y"]
```

```python
# plot roc curves 

plt.figure(figsize = (20,12))

for i, model in enumerate(estimators):
    plt.subplot(2,3, i+1)
    
    plt.xlabel("False positive rate")
    plt.ylabel("True positive rate")
    
    # plot parameters using unscaled predictors
    params = roc_params(estimators[i], X_train, X_test)
    plt.plot(params[0], params[1], colors[i] + "--", linewidth = 2, label = "Unscaled")
    
    # plot parameters using scaled predictors
    scaled_params = roc_params(estimators[i], X_train_std, X_test_std)
    plt.plot(scaled_params[0], scaled_params[1], colors[i], linewidth = 2, label = "Scaled")
    plt.legend(loc = 4)
    plt.title(names[i]+" "+str(round(scaled_params[2],2))+"vs"+str(round(params[2],2)), fontsize = 20)
plt.tight_layout()
plt.show()
```

```python
scores = [accuracy_score, precision_score, recall_score, f1_score]

def get_metrics(X_train, X_test):
    
    # create empty lists
    LogisticReg = []
    SVMC = []
    GaussNB = []
    DecisionTree = []
    RandomForest = []
    kNN9 = []
    
    # list of lists
    lists = [LogisticReg, SVMC, GaussNB, DecisionTree, RandomForest, kNN9]
    
    # populate lists with scores of each scoring method
    for i, model in enumerate(estimators):
        for score in scores:
            estimator = model
            estimator.fit(X_train, y_train)
            pred = estimator.predict(X_test)
            lists[i].append(score(y_test, pred))
        
    # create a dataframe which aggregates the lists
    scores_df = pd.DataFrame(data = [LogisticReg, SVMC, GaussNB, DecisionTree, RandomForest, kNN9])
    scores_df.index = ["LogisticReg", "SVMC", "GaussNB", "DecisionTree", "RandomForest", "kNN9"]
    scores_df.columns = ["Accuracy", "Precision", "Recall", "F1"]
    return scores_df
```

```python
print("unscaled:")
get_metrics(X_train, X_test)
```

```python
print("scaled:")
get_metrics(X_train_std, X_test_std)
```
